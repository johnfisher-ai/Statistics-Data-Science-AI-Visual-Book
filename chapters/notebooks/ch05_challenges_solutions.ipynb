<div style="color:#4a5578;margin-top:6px">Give the level (Nominal, Ordinal, Interval, or Ratio) for: blood type, exam grade (A–F), temperature in °C, weight in kg, finishing place (1st/2nd/3rd), temperature in Kelvin, calendar year.</div>
</div>

```python
answer = pd.DataFrame({
  "variable":["blood type","exam grade (A-F)","temperature °C","weight kg",
              "finishing place","temperature Kelvin","calendar year"],
  "level":   ["Nominal","Ordinal","Interval","Ratio","Ordinal","Ratio","Interval"],
  "why":     ["names, no order","ordered grades, unequal gaps","equal gaps, no true zero",
              "true zero (0 kg = none)","ranked order only","0 K = no heat (true zero)",
              "equal gaps, year 0 is arbitrary"],
})
answer
```

**Answer:** the two tricky ones — **Kelvin is ratio** (0 K is a true "no heat" zero, so 200 K really is twice 100 K), while **°C and calendar years are interval** (their zeros are arbitrary).

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 2 · TRUE ZERO</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Which support "twice as much"?</div>
<div style="color:#4a5578;margin-top:6px">Of these, which allow valid "twice as much" statements (i.e., are ratio scales)? temperature °C, height, income, IQ score.</div>
</div>

```python
check = pd.DataFrame({
  "variable":["temperature °C","height","income","IQ score"],
  "ratio scale?":["No","Yes","Yes","No"],
  "\"twice\" valid?":["No (no true zero)","Yes","Yes (0 income = none)","No (no true zero)"],
})
check
```

**Answer:** **Height** and **income** are ratio (true zero), so "twice as tall / twice the income" make sense. **°C** and **IQ** are interval — their zeros are arbitrary, so "twice the temperature / twice as smart" are meaningless.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#d97706;letter-spacing:1px">CHALLENGE 3 · RIGHT CENTER</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Pick the valid average</div>
<div style="color:#4a5578;margin-top:6px">Which measure of center (mode, median, or mean) is most appropriate for: favorite color, satisfaction rating (Low/Med/High), and annual salary?</div>
</div>

```python
colors  = pd.Series(["red","blue","red","green","red"])    # nominal
rating  = pd.Series([1, 3, 2, 3, 1])                       # ordinal (Low=1..High=3)
salary  = pd.Series([42000, 55000, 38000, 61000, 50000])   # ratio

print("Favorite color (nominal)   -> mode  :", colors.mode()[0])
print("Satisfaction  (ordinal)    -> median:", int(rating.median()))
print(f"Annual salary (ratio)      -> mean  : {salary.mean():,.0f}")
```

**Answer:** Nominal → **mode** (most frequent). Ordinal → **median** (mean of codes is misleading). Ratio → **mean** is fully valid.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#2563eb;letter-spacing:1px">CHALLENGE 4 · SPOT THE ERROR</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">What went wrong?</div>
<div style="color:#4a5578;margin-top:6px">An analyst reports the "average" zip code as 60412 and the "mean satisfaction" as 3.4 from labels coded Low=1…High=5. What's wrong, and what should they do instead?</div>
</div>

```python
# Zip codes are NOMINAL labels — averaging them is meaningless
zips = pd.Series([60201, 60601, 60412, 60634])
print("Zip code is NOMINAL -> the average has no meaning.")
print("  Better: report the mode or counts per zip:", dict(zips.value_counts()))

# Satisfaction is ORDINAL -> mean assumes equal gaps; use the median
sat = pd.Series([1,5,3,5,1])
print(f"\nSatisfaction is ORDINAL -> report median ({sat.median():.0f}), not the mean ({sat.mean():.1f}).")
```

**Answer:** A *zip code* is a **nominal** label, so its mean is nonsense (use mode/counts). *Satisfaction* is **ordinal**, so the mean assumes equal spacing between levels — report the **median** instead. Matching the statistic to the level is the whole point of this chapter.
