<span style="font-size:13px;font-weight:800;color:#0891b2;letter-spacing:1px">CHALLENGE 1 · POPULATION vs SAMPLE</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Identify each part</div>
<div style="color:#4a5578;margin-top:6px">A university has 20,000 students. A researcher surveys 500 of them about study habits. Name the population, the sample, and the sample size n.</div>
</div>

```python
answer = {
  "Population": "all 20,000 students at the university",
  "Sample":     "the 500 students who were surveyed",
  "Sample size n": 500,
}
for k, v in answer.items():
    print(f"{k:<15}: {v}")
```

**Answer:** Population = the entire group of interest (20,000 students); sample = the subset actually measured (500); n = 500. We study the sample to learn about the population.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 2 · PARAMETER or STATISTIC</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Classify and give the symbol</div>
<div style="color:#4a5578;margin-top:6px">Label each as a parameter or a statistic, and give its symbol: (a) the mean height of ALL NBA players; (b) the mean height of 30 sampled players; (c) the proportion of a 1,000-person poll who approve.</div>
</div>

```python
rows = pd.DataFrame({
  "quantity":["mean height of ALL NBA players","mean height of 30 sampled players","approval in a 1,000-person poll"],
  "type":    ["Parameter","Statistic","Statistic"],
  "symbol":  ["μ (mu)","x̄ (x-bar)","p̂ (p-hat)"],
})
print(rows.to_string(index=False))
```

**Answer:** Anything describing the **whole population** is a parameter (Greek letters: μ, σ, p). Anything from a **sample** is a statistic (Latin: x̄, s, p̂). "All players" → parameter; "sampled / polled" → statistic.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#d97706;letter-spacing:1px">CHALLENGE 3 · COMPUTE A STATISTIC</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Estimate the parameter from a sample</div>
<div style="color:#4a5578;margin-top:6px">A sample of 8 customers spent (in $): [12, 18, 9, 22, 15, 30, 11, 19]. Compute the sample mean x̄ and sample proportion p̂ who spent more than $15. What are these estimating?</div>
</div>

```python
spend = pd.Series([12, 18, 9, 22, 15, 30, 11, 19])
xbar = spend.mean()
phat = (spend > 15).mean()
print(f"Sample mean       x̄ = ${xbar:.2f}")
print(f"Sample proportion p̂ = {phat:.2%} spent more than $15")
print("\nThese sample STATISTICS estimate the population PARAMETERS")
print("(μ = mean spend of ALL customers, p = proportion of ALL customers over $15).")
```

**Answer:** x̄ = **$17.00**, p̂ = **50%**. They're our best estimates of the unknown population mean (μ) and proportion (p) of *all* customers.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#2563eb;letter-spacing:1px">CHALLENGE 4 · CENSUS or SAMPLING</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Pick the right approach</div>
<div style="color:#4a5578;margin-top:6px">For each, say whether a CENSUS or a SAMPLE is more appropriate, and why: (a) a national population count every 10 years; (b) testing whether a factory's light bulbs last 1,000 hours; (c) a quick exit poll on election night.</div>
</div>

```python
decisions = {
  "National population count": "CENSUS — exact counts of everyone are legally/strategically required",
  "Light-bulb lifetime test":  "SAMPLE — testing is destructive; you can't burn out every bulb",
  "Election-night exit poll":   "SAMPLE — results are needed fast and cheaply, with a margin of error",
}
for k, v in decisions.items():
    print(f"• {k}\n    -> {v}\n")
```

**Answer:** Use a **census** when you truly need every individual (national count). Use a **sample** when a census is impossible (destructive testing), too slow, or too costly (exit polls) — accepting a small, quantified margin of error.
