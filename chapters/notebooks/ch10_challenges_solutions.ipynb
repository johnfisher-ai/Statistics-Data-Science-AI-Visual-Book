## ⚙️ Setup

```python
import numpy as np
import pandas as pd
data = pd.Series([55, 60, 62, 68, 70, 72, 75, 80, 85, 90])
print("Dataset:", list(data))
```

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0891b2;letter-spacing:1px">CHALLENGE 1 · PERCENTILE RANK</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Where does a value stand?</div>
<div style="color:#4a5578;margin-top:6px">Using the dataset above, what is the percentile RANK of the score 70 (the percent scoring at or below it)?</div>
</div>

```python
x = 70
rank = (data <= x).mean() * 100
print(f"Percentile rank of {x} = {rank:.0f}%")
```

**Answer:** 5 of the 10 values (55, 60, 62, 68, 70) are ≤ 70, so the percentile rank is **50%**. About half the scores are at or below 70.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 2 · PERCENTILE VALUE</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Go the other direction</div>
<div style="color:#4a5578;margin-top:6px">What score sits at the 80th percentile of the dataset?</div>
</div>

```python
p80 = np.percentile(data, 80)
print(f"80th percentile = {p80:.1f}")
```

**Answer:** The 80th percentile is about **84** (NumPy's linear method interpolates between 80 and 85). Roughly 80% of scores fall at or below it.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#d97706;letter-spacing:1px">CHALLENGE 3 · QUARTILES & DECILES</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Cut the data up</div>
<div style="color:#4a5578;margin-top:6px">Find Q1, the median (Q2), Q3, and the 3rd decile (30th percentile) of the dataset.</div>
</div>

```python
q1, q2, q3 = np.percentile(data, [25, 50, 75])
d3 = np.percentile(data, 30)
print(f"Q1 = {q1:.1f}, Median (Q2) = {q2:.1f}, Q3 = {q3:.1f}")
print(f"3rd decile (30th percentile) = {d3:.1f}")
```

**Answer:** Q1 ≈ **64.5**, median = **71**, Q3 ≈ **78.75**, and the 3rd decile ≈ **66.6**. Quartiles and deciles are just percentiles at the 25/50/75 and 10/20/...90 marks.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#059669;letter-spacing:1px">CHALLENGE 4 · INTERPRETATION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Read a percentile correctly</div>
<div style="color:#4a5578;margin-top:6px">A pediatrician says a toddler is in the 30th percentile for height. What does that mean, and is it a problem?</div>
</div>

```python
msg = ("About 30% of children that age are the same height or shorter, ",
       "and about 70% are taller. It is NOT a grade or a score of 30%. ",
       "A single percentile is not \"good\" or \"bad\"; clinicians watch the trend over time.")
print("".join(msg))
```

**Answer:** ~30% of same-age children are that height or shorter (≈70% are taller). It is a *ranking*, not a 30% grade, and a stable 30th-percentile child is perfectly healthy. The trend across visits matters more than any one number.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#2563eb;letter-spacing:1px">CHALLENGE 5 · METHODS DIFFER</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Two valid answers</div>
<div style="color:#4a5578;margin-top:6px">For the small set [3, 6, 9, 12], compute the first quartile (25th percentile) using NumPy's "lower" method and its default "linear" method. Why do they differ?</div>
</div>

```python
small = np.array([3, 6, 9, 12])
print(f"method=lower  -> {np.percentile(small, 25, method='lower'):.2f}")
print(f"method=linear -> {np.percentile(small, 25, method='linear'):.2f}")
```

**Answer:** "lower" returns **3** (it drops to the nearest data value below), while "linear" interpolates to **5.25** (three-quarters of the way from 3 to 6). The 25% mark falls between two data points, so each method resolves it differently. Neither is wrong, but always report the method you used.
