## ⚙️ Setup

```python
import numpy as np
import pandas as pd
from scipy import stats
rng = np.random.default_rng(111)
print("Ready.")
```

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0891b2;letter-spacing:1px">CHALLENGE 1 · NAME THE SKEW</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Read direction from center</div>
<div style="color:#4a5578;margin-top:6px">A dataset of house prices has mean 420,000 and median 350,000. Is it left-skewed or right-skewed, and which single number better describes a typical home?</div>
</div>

```python
mean_, median_ = 420_000, 350_000
gap = mean_ - median_
print(f"mean - median = {gap:,}  ->", "right-skewed" if gap>0 else "left-skewed")
```

**Answer:** The mean sits **above** the median, so the long tail points right: the data is **right-skewed** (a few expensive homes pull the mean up). The **median (350,000)** better describes a typical home, because it is not dragged by the high outliers.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 2 · MEASURE THE SKEW</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Compute and classify</div>
<div style="color:#4a5578;margin-top:6px">Generate a right-skewed sample with rng.exponential(scale=5, size=3000). Compute its skewness with scipy and classify it using the |skew| < 0.5 / 0.5-1 / >1 rule of thumb.</div>
</div>

```python
sample = rng.exponential(scale=5, size=3000)
sk = stats.skew(sample)
band = "fairly symmetric" if abs(sk)<0.5 else ("moderately skewed" if abs(sk)<1 else "highly skewed")
print(f"skewness = {sk:+.2f}  ->  {band}, tail points {'right' if sk>0 else 'left'}")
```

**Answer:** An exponential sample has true skewness 2, so the computed value is **well above 1**: it is **highly skewed** to the **right**. The sign gives direction (+ = right), the magnitude gives severity. Remember the 0.5 and 1 cutoffs are heuristics, not hard law.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#d97706;letter-spacing:1px">CHALLENGE 3 · KURTOSIS MEANS TAILS</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Classify and interpret</div>
<div style="color:#4a5578;margin-top:6px">For a normal sample and a Laplace (double-exponential) sample of the same size, compute EXCESS kurtosis with scipy. Label each mesokurtic / leptokurtic / platykurtic and say what it implies.</div>
</div>

```python
normal_s = rng.normal(0, 1, 5000)
laplace_s = rng.laplace(0, 1, 5000)
def label(k): return "mesokurtic (~normal tails)" if abs(k)<0.5 else ("leptokurtic (heavy tails)" if k>0 else "platykurtic (light tails)")
for name,d in [("normal",normal_s),("laplace",laplace_s)]:
    k = stats.kurtosis(d)   # excess: normal ≈ 0
    print(f"{name:8} excess kurtosis = {k:+.2f}  ->  {label(k)}")
```

**Answer:** The normal sample lands near **0 (mesokurtic)**; the Laplace sample is **clearly positive (leptokurtic)**. Positive excess kurtosis means **heavier tails, more outliers**, not a taller peak. That is the modern, correct reading of kurtosis: it is about tail extremity, not peakedness.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#059669;letter-spacing:1px">CHALLENGE 4 · TWO PEAKS</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Interpret a bimodal shape</div>
<div style="color:#4a5578;margin-top:6px">A fitness app plots daily step counts for all users and sees two clear peaks, one around 3,000 steps and one around 11,000. What does this shape most likely indicate, and what should the analyst do?</div>
</div>

```python
msg = ("Two peaks suggest two subpopulations mixed together, ",
       "for example mostly-sedentary users and active users. ",
       "A single overall average would fall in the empty valley ",
       "between them and describe almost nobody. ",
       "The analyst should segment the users and analyze each group separately.")
print("".join(msg))
```

**Answer:** A **bimodal** shape almost always signals **two mixed subpopulations** (here, sedentary vs. active users). The overall mean lands in the valley between the peaks and represents no real user. The right move is to **disaggregate** and study each group on its own.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#2563eb;letter-spacing:1px">CHALLENGE 5 · TAME THE TAIL</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Apply a log transform</div>
<div style="color:#4a5578;margin-top:6px">Take a strongly right-skewed sample rng.lognormal(mean=2, sigma=1, size=4000). Report its skewness, apply a log transform, and report the skewness again. Explain what changed and one caveat.</div>
</div>

```python
raw = rng.lognormal(mean=2, sigma=1, size=4000)
logged = np.log(raw)
print(f"raw skewness     = {stats.skew(raw):+.2f}")
print(f"log() skewness   = {stats.skew(logged):+.2f}")
```

**Answer:** The raw data is **strongly right-skewed** (skew well above 1); after `np.log` the skewness drops to **near 0**, the long tail is pulled in and the shape is nearly symmetric. Caveat: the log is **undefined for 0 or negative values** (use `log1p` or a shift), and results are now on a **log scale**, so interpret them accordingly.
