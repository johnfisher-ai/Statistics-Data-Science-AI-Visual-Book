## ⚙️ Setup

```python
import numpy as np
from scipy import stats
rng = np.random.default_rng(401)
print("ready")
```

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 1 · STANDARD ERROR</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">sigma over root n</div>
<div style="color:#4a5578;margin-top:6px">A population has standard deviation 20. Find the standard error of the sample mean for n = 100 and for n = 400.</div>
</div>

```python
sigma = 20
for n in [100, 400]:
    print(f"n={n}: SE = 20/sqrt({n}) = {sigma/np.sqrt(n):.2f}")
```

**Answer:** SE = &#963;/&#8730;n, so for n = 100, SE = 20/10 = **2.0**, and for n = 400, SE = 20/20 = **1.0**. Quadrupling the sample halved the standard error, the square-root law in action.

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0891b2;letter-spacing:1px">CHALLENGE 2 · CLT IN ACTION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">A probability about a mean</div>
<div style="color:#4a5578;margin-top:6px">Daily sales have mean 500 and sd 120 (skewed). For a sample of n = 36 days, find P(sample mean > 540), using the CLT.</div>
</div>

```python
mu, sd, n = 500, 120, 36
SE = sd/np.sqrt(n)
z = (540 - mu)/SE
print(f"SE = {SE:.2f}, z = (540-500)/SE = {z:.2f}")
print(f"P(sample mean > 540) = {1 - stats.norm.cdf(z):.4f}")
```

**Answer:** the CLT makes the sample mean approximately normal with SE = 120/6 = 20. Then z = (540 &minus; 500)/20 = 2.0, so P(mean &gt; 540) = 1 &minus; &Phi;(2) &approx; **0.0228**. Even though daily sales are skewed, the *mean* of 36 days is nearly normal.

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 3 · THE sqrt(N) LAW</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">How much more data?</div>
<div style="color:#4a5578;margin-top:6px">Your current sample gives a standard error of 4. You want to cut it to 1. By what factor must you increase the sample size?</div>
</div>

```python
# SE scales like 1/sqrt(n); to divide SE by 4, multiply n by 4^2
factor = (4/1)**2
print(f"to reduce SE by a factor of 4, increase n by {factor:.0f}x")
```

**Answer:** since SE &prop; 1/&#8730;n, dividing the SE by 4 requires multiplying n by 4&sup2; = **16**. Precision improves only with the square root of sample size, so each extra digit of accuracy is dramatically more expensive.

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#db2777;letter-spacing:1px">CHALLENGE 4 · CONFIDENCE INTERVAL</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Mean plus or minus a margin</div>
<div style="color:#4a5578;margin-top:6px">A sample of 64 has mean 50 and the population sd is 16. Build an approximate 95% confidence interval for the true mean.</div>
</div>

```python
xbar, sigma, n = 50, 16, 64
SE = sigma/np.sqrt(n)
lo, hi = xbar - 1.96*SE, xbar + 1.96*SE
print(f"SE = 16/8 = {SE:.1f}")
print(f"95% CI = 50 +/- 1.96*{SE:.1f} = [{lo:.2f}, {hi:.2f}]")
```

**Answer:** SE = 16/&#8730;64 = 2.0, so the 95% interval is 50 &plusmn; 1.96(2) = **[46.08, 53.92]**. We are about 95% confident the true mean lies in this range, an interval built directly on the CLT.

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0f766e;letter-spacing:1px">CHALLENGE 5 · A POLL'S MARGIN</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Standard error of a proportion</div>
<div style="color:#4a5578;margin-top:6px">A poll of 1,000 people finds 60% support. Find the standard error of the proportion and the 95% margin of error.</div>
</div>

```python
p, n = 0.60, 1000
SE = np.sqrt(p*(1-p)/n)
print(f"SE = sqrt(0.6*0.4/1000) = {SE:.4f}")
print(f"95% margin of error = 1.96*SE = +/- {1.96*SE:.4f}  (about +/- 3 points)")
```

**Answer:** SE = &#8730;(0.6&middot;0.4/1000) &approx; **0.0155**, so the 95% margin of error is 1.96 &times; 0.0155 &approx; **&plusmn;0.030**, the familiar "&plusmn;3 percentage points" of political polls. A proportion is just a mean of 0s and 1s, so the same CLT machinery applies.
