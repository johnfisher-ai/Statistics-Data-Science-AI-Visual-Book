## ⚙️ Setup

```python
import numpy as np
from scipy import stats
rng = np.random.default_rng(431)
print("ready")
```

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 1 · LINEAR TRANSFORM</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Convert the units</div>
<div style="color:#4a5578;margin-top:6px">Temperatures in Celsius are Normal(20, 5). Fahrenheit is F = 1.8*C + 32. Find the mean and standard deviation of F.</div>
</div>

```python
mu_c, sd_c = 20, 5
mu_f = 1.8*mu_c + 32
sd_f = abs(1.8)*sd_c
print(f"mean F = 1.8*20 + 32 = {mu_f}")
print(f"sd F   = 1.8*5      = {sd_f}")
```

**Answer:** mean = 1.8(20) + 32 = **68&deg;F**; sd = 1.8(5) = **9&deg;F**. The +32 shifts the mean but not the spread; only the multiplier 1.8 scales the standard deviation. F is still normal.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#6d28d9;letter-spacing:1px">CHALLENGE 2 · NONLINEAR TRANSFORM</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Make a lognormal</div>
<div style="color:#4a5578;margin-top:6px">Let X ~ Normal(0, 1) and Y = exp(X). Simulate and confirm Y is right-skewed with the lognormal's median of 1.</div>
</div>

```python
X = rng.normal(0,1,size=300_000); Y = np.exp(X)
print(f"skew of Y = {stats.skew(Y):.3f} (positive -> right-skewed)")
print(f"median of Y = {np.median(Y):.3f} (lognormal median = exp(0) = 1)")
```

**Answer:** Y is strongly right-skewed (positive skew) with median &approx; **1.0**, since the median maps through exp as exp(0) = 1. The exponential transform turns a symmetric normal into the asymmetric lognormal.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 3 · CHANGE OF VARIABLES</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Apply the Jacobian</div>
<div style="color:#4a5578;margin-top:6px">For Y = X^2 with X ~ Uniform(0,1), the density of Y is f_Y(y) = 1/(2*sqrt(y)). Verify by simulation that more mass piles up near 0.</div>
</div>

```python
X = rng.random(300_000); Y = X**2
# fraction of Y below 0.25 should be sqrt(0.25) = 0.5 (since CDF of Y is sqrt(y))
print(f"P(Y < 0.25) simulated = {np.mean(Y<0.25):.3f}  (theory sqrt(0.25) = 0.5)")
print(f"P(Y < 0.04) simulated = {np.mean(Y<0.04):.3f}  (theory sqrt(0.04) = 0.2)")
```

**Answer:** the CDF of Y is &#8730;y, so P(Y &lt; 0.25) = &#8730;0.25 = **0.5** and P(Y &lt; 0.04) = 0.2, both matched by simulation. Squaring compresses the interval near 0, so the density f_Y(y) = 1/(2&#8730;y) blows up there, exactly the Jacobian at work.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#db2777;letter-spacing:1px">CHALLENGE 4 · INVERSE-TRANSFORM SAMPLING</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Build a sampler</div>
<div style="color:#4a5578;margin-top:6px">Use inverse-transform sampling to generate Exponential(rate=2) variates from uniforms, and check the mean (which should be 1/2).</div>
</div>

```python
U = rng.random(300_000)
rate = 2
X = -np.log(1 - U) / rate          # inverse CDF of exponential(rate)
print(f"sample mean = {X.mean():.4f}  (theory 1/rate = {1/rate})")
```

**Answer:** inverting the exponential CDF gives X = &minus;ln(1&minus;U)/rate, and the sample mean is about **0.5**, matching 1/rate. Any distribution with an invertible CDF can be sampled this way, the foundation of random-variate generation.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#5b21b6;letter-spacing:1px">CHALLENGE 5 · REPARAMETERIZATION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Differentiable sampling</div>
<div style="color:#4a5578;margin-top:6px">Write a sample from Normal(mu=10, sigma=3) as a transform of standard-normal noise, the reparameterization trick, and confirm the statistics.</div>
</div>

```python
mu, sigma = 10, 3
eps = rng.normal(0,1,size=300_000)
z = mu + sigma*eps
print(f"reparameterized z: mean {z.mean():.3f} (target {mu}), sd {z.std():.3f} (target {sigma})")
```

**Answer:** z = mu + sigma&middot;epsilon recovers Normal(10, 3) exactly. Because z is a differentiable function of mu and sigma, gradients pass through the sampling step, which is what makes variational autoencoders trainable by backpropagation.
