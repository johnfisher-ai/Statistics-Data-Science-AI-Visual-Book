## ⚙️ Setup

```python
import numpy as np
rng = np.random.default_rng(441)
print("ready")
```

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 1 · CONDITIONAL MEAN</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Average within a group</div>
<div style="color:#4a5578;margin-top:6px">A factory has two machines. Machine A makes parts with mean weight 100g (sd 5); machine B mean 110g (sd 5). Simulate and estimate E[weight | machine].</div>
</div>

```python
m = rng.integers(0,2,size=40_000)
w = rng.normal(np.where(m==0,100,110), 5)
print(f"E[weight | A] = {w[m==0].mean():.2f}  (true 100)")
print(f"E[weight | B] = {w[m==1].mean():.2f}  (true 110)")
```

**Answer:** the conditional expectation is about **100g** given machine A and **110g** given machine B. E[Y | X] is a function of X: one average per machine, not a single overall number.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#6d28d9;letter-spacing:1px">CHALLENGE 2 · TOWER PROPERTY</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Recover the overall mean</div>
<div style="color:#4a5578;margin-top:6px">Using the same data, verify E[E[weight|machine]] = E[weight] by averaging the two conditional means, weighted by how often each machine is used.</div>
</div>

```python
w_A, w_B = (m==0).mean(), (m==1).mean()
tower = w_A*w[m==0].mean() + w_B*w[m==1].mean()
print(f"E[weight] directly = {w.mean():.3f}")
print(f"tower E[E[W|M]]    = {tower:.3f}")
```

**Answer:** both equal about **105g**. The tower property recovers the overall mean by averaging the within-machine means, weighted by usage. Because each machine runs about half the time, the overall mean sits near the midpoint of 100 and 110.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 3 · BEST PREDICTOR</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Conditional mean beats the global mean</div>
<div style="color:#4a5578;margin-top:6px">Show that predicting each part's weight with its machine's conditional mean gives a lower mean squared error than always predicting the overall mean.</div>
</div>

```python
pred_cond = np.where(m==0, w[m==0].mean(), w[m==1].mean())
pred_glob = np.full_like(w, w.mean())
print(f"MSE with E[W|M]    = {((w-pred_cond)**2).mean():.2f}")
print(f"MSE with E[W] only = {((w-pred_glob)**2).mean():.2f}")
```

**Answer:** conditioning on the machine lowers the MSE (about 25, the within-group variance 5&#178;) versus ignoring it (about 50). No function of the machine label can beat E[W | machine] on squared error, which is exactly why it is "the best predictor".

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#db2777;letter-spacing:1px">CHALLENGE 4 · TOTAL VARIANCE</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Within plus between</div>
<div style="color:#4a5578;margin-top:6px">Decompose the total variance of weight into E[Var(W|M)] (within) and Var(E[W|M]) (between), and check they sum to Var(W).</div>
</div>

```python
wts = np.array([(m==0).mean(),(m==1).mean()])
cm  = np.array([w[m==0].mean(), w[m==1].mean()])
within  = sum(wts[k]*w[m==k].var() for k in range(2))
between = (wts*(cm - w.mean())**2).sum()
print(f"Var(W) total = {w.var():.2f}")
print(f"within {within:.2f} + between {between:.2f} = {within+between:.2f}")
```

**Answer:** the within-machine variance (&approx; 25, from the sd of 5) plus the between-machine variance (&approx; 25, from the 10g gap split across two equal groups) sums to the total variance of weight. This law of total variance underlies ANOVA and bias-variance analysis.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#5b21b6;letter-spacing:1px">CHALLENGE 5 · REGRESSION AS E[Y|X]</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Recover a curve</div>
<div style="color:#4a5578;margin-top:6px">Generate y = x^2 + noise and estimate E[Y | X] by binning. Confirm the estimated conditional mean near x=2 is close to 4.</div>
</div>

```python
x = rng.uniform(0,3,size=5000); y = x**2 + rng.normal(0,0.5,size=5000)
near2 = (x>1.9)&(x<2.1)
print(f"estimated E[Y | X near 2] = {y[near2].mean():.3f}  (true 2^2 = 4)")
```

**Answer:** the estimated conditional mean near x = 2 is about **4.0**, matching the true E[Y | X] = x&#178;. Regression of any kind, linear, tree-based, or neural, is an estimator of this conditional-expectation function.
