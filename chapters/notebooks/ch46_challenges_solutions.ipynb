## ⚙️ Setup

```python
import numpy as np
rng = np.random.default_rng(461)
print("ready")
```

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 1 · UNBIASED?</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Check the sample mean</div>
<div style="color:#4a5578;margin-top:6px">Is the sample mean an unbiased estimator of the population mean? Simulate samples of size 20 from Normal(50, 10) and check.</div>
</div>

```python
ests = np.array([rng.normal(50,10,size=20).mean() for _ in range(100_000)])
print(f"E[sample mean] = {ests.mean():.3f}  (true 50)")
print(f"bias = {ests.mean()-50:+.3f}  -> unbiased")
```

**Answer:** the average of the estimates is **50**, so the bias is essentially 0: the sample mean is unbiased. On average it neither over- nor under-shoots the true mean, whatever the sample size.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#6d28d9;letter-spacing:1px">CHALLENGE 2 · BESSEL'S CORRECTION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">n versus n-1</div>
<div style="color:#4a5578;margin-top:6px">For samples of size 4 from Normal(0, 3) (variance 9), compare the variance estimator dividing by n to the one dividing by n-1.</div>
</div>

```python
n=4
var_n   = np.mean([rng.normal(0,3,size=n).var(ddof=0) for _ in range(200_000)])
var_nm1 = np.mean([rng.normal(0,3,size=n).var(ddof=1) for _ in range(200_000)])
print(f"true variance = 9")
print(f"E[/n]    = {var_n:.3f}  (biased low; factor (n-1)/n = {(n-1)/n})")
print(f"E[/(n-1)]= {var_nm1:.3f}  (unbiased)")
```

**Answer:** dividing by n averages to about **6.75** (= 9 &times; 3/4), biased low, while dividing by n &#8722; 1 averages to **9**, unbiased. Bessel's correction (n &#8722; 1) compensates for using the sample mean in place of the unknown true mean.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 3 · MLE</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Estimate a rate</div>
<div style="color:#4a5578;margin-top:6px">Derive and compute the maximum-likelihood estimate of the rate of an exponential from a sample (true rate 0.4).</div>
</div>

```python
data = rng.exponential(1/0.4, size=5000)
mle = 1/data.mean()
print(f"MLE rate = 1/mean = {mle:.3f}  (true 0.4)")
```

**Answer:** the exponential MLE is rate = 1/mean &approx; **0.4**. Maximizing the log-likelihood n&middot;log(&#955;) &#8722; &#955;&middot;&Sigma;x gives &#955; = 1/x&#772; exactly, a textbook maximum-likelihood result.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#db2777;letter-spacing:1px">CHALLENGE 4 · CONSISTENCY</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Error shrinks with n</div>
<div style="color:#4a5578;margin-top:6px">Show the MLE of a normal mean becomes more accurate as n grows from 10 to 10,000.</div>
</div>

```python
for n in [10, 100, 1000, 10000]:
    ests = [rng.normal(5,2,size=n).mean() for _ in range(2000)]
    print(f"n={n:6d}: sd of estimate = {np.std(ests):.4f}  (theory 2/sqrt(n) = {2/np.sqrt(n):.4f})")
```

**Answer:** the spread of the estimate shrinks like 2/&#8730;n, from about 0.63 at n = 10 to 0.02 at n = 10,000, converging on the true mean. This is **consistency**: with enough data the estimator is essentially certain to be close.

<div style="background:#f5f3ff;border-left:5px solid #7c3aed;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#5b21b6;letter-spacing:1px">CHALLENGE 5 · MSE DECOMPOSITION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Bias squared plus variance</div>
<div style="color:#4a5578;margin-top:6px">For a deliberately biased estimator (the sample mean shrunk by 10%), compute its bias, variance, and verify MSE = bias^2 + variance.</div>
</div>

```python
true_mu = 20
ests = np.array([0.9*rng.normal(true_mu,4,size=50).mean() for _ in range(100_000)])
bias = ests.mean() - true_mu
var  = ests.var()
mse  = ((ests - true_mu)**2).mean()
print(f"bias = {bias:.3f}, bias^2 = {bias**2:.3f}")
print(f"variance = {var:.3f}")
print(f"bias^2 + variance = {bias**2+var:.3f}   vs   MSE = {mse:.3f}")
```

**Answer:** shrinking the mean by 10% introduces a bias of about &#8722;2 (bias&#178; &approx; 4), and the identity holds: **MSE = bias&#178; + variance**. This decomposition is the exact form of the bias-variance trade-off, sometimes a little bias buys a large drop in variance and a lower overall MSE, which is the whole rationale for regularization.
