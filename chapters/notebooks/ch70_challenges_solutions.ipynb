## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
rng = np.random.default_rng(700)
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">z-interval (sigma known)</div>
<div style="color:#4a5578;margin-top:5px">Sigma is known to be 8. From a sample of 40 from Normal(50, 8), build a 95% z-interval for the mean.</div>
</div>

```python
s = rng.normal(50, 8, 40); xb=s.mean(); se=8/np.sqrt(40); z=stats.norm.ppf(0.975)
print(f"95% z-interval: [{xb-z*se:.2f}, {xb+z*se:.2f}]")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">t-interval (sigma unknown)</div>
<div style="color:#4a5578;margin-top:5px">Now pretend sigma is unknown: use the sample sd and the t-distribution for the same sample.</div>
</div>

```python
sd = s.std(ddof=1); se=sd/np.sqrt(40)
lo,hi = stats.t.interval(0.95, df=39, loc=s.mean(), scale=se)
print(f"95% t-interval: [{lo:.2f}, {hi:.2f}]  (t* = {stats.t.ppf(0.975,39):.3f} vs z* {stats.norm.ppf(0.975):.3f})")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">t* shrinks toward z*</div>
<div style="color:#4a5578;margin-top:5px">Tabulate the 95% t critical value for df = 4, 9, 29, 99 and compare to z* = 1.96.</div>
</div>

```python
for d in [4,9,29,99]: print(f"df={d:>3}: t* = {stats.t.ppf(0.975,d):.3f}")
print(f"z* = {stats.norm.ppf(0.975):.3f}  (t* -> z* as df grows)")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Coverage check</div>
<div style="color:#4a5578;margin-top:5px">From a skewed Gamma(3,5) population, confirm the t-interval coverage is near 95% at n=40.</div>
</div>

```python
pop=rng.gamma(3,5,800_000); MU=pop.mean()
def cov(n=40):
    x=rng.choice(pop,n,replace=False); se=x.std(ddof=1)/np.sqrt(n)
    lo,hi=stats.t.interval(0.95,n-1,x.mean(),se); return lo<=MU<=hi
print(f"coverage at n=40: {np.mean([cov() for _ in range(4000)])*100:.1f}%")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: mean delivery time</div>
<div style="color:#4a5578;margin-top:5px">Load ch70_delivery_times.xlsx and report the 95% t-interval for the mean delivery time.</div>
</div>

```python
try:    ship = pd.read_excel("../../data/ch70_delivery_times.xlsx", sheet_name="Shipments")
except FileNotFoundError: ship = pd.read_excel(BASE+"ch70_delivery_times.xlsx", sheet_name="Shipments")
h=ship.delivery_hours; n=len(h); se=h.std(ddof=1)/np.sqrt(n)
lo,hi=stats.t.interval(0.95,n-1,h.mean(),se)
print(f"n={n}: mean {h.mean():.2f} h, 95% t-interval [{lo:.2f}, {hi:.2f}] h")
```
