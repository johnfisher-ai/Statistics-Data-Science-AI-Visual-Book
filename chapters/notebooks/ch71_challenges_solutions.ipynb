## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
z = stats.norm.ppf(0.975); rng = np.random.default_rng(710)
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">CI for a proportion</div>
<div style="color:#4a5578;margin-top:5px">220 of 1000 surveyed customers churned. Build a 95% CI for the churn rate.</div>
</div>

```python
succ, n = 220, 1000; p=succ/n; se=np.sqrt(p*(1-p)/n)
print(f"p-hat={p:.3f}, 95% CI [{p-z*se:.3f}, {p+z*se:.3f}]")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Success-failure condition</div>
<div style="color:#4a5578;margin-top:5px">For 6 successes out of 40, check whether the normal-approx CI is appropriate.</div>
</div>

```python
succ,n=6,40
print(f"successes={succ}, failures={n-succ}; both >= 10? {succ>=10 and n-succ>=10}")
print("-> normal approx is shaky here; prefer the Wilson/exact interval")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Difference of two proportions</div>
<div style="color:#4a5578;margin-top:5px">Group A: 90/600 converted. Group B: 120/620. Build a 95% CI for pB - pA.</div>
</div>

```python
pA,nA=90/600,600; pB,nB=120/620,620; diff=pB-pA
se=np.sqrt(pA*(1-pA)/nA + pB*(1-pB)/nB)
print(f"diff={diff:+.3f}, 95% CI [{diff-z*se:+.3f}, {diff+z*se:+.3f}]")
print("excludes 0?", (diff-z*se>0) or (diff+z*se<0))
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Difference of two means</div>
<div style="color:#4a5578;margin-top:5px">Two groups spend Normal(40,12) n=200 and Normal(44,13) n=210. Build a 95% CI for the difference of means.</div>
</div>

```python
a=rng.normal(40,12,200); b=rng.normal(44,13,210); diff=b.mean()-a.mean()
se=np.sqrt(a.var(ddof=1)/len(a)+b.var(ddof=1)/len(b))
dfw=se**4/((a.var(ddof=1)/len(a))**2/(len(a)-1)+(b.var(ddof=1)/len(b))**2/(len(b)-1))
t=stats.t.ppf(0.975,dfw)
print(f"diff={diff:+.2f}, 95% CI [{diff-t*se:+.2f}, {diff+t*se:+.2f}]")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: A/B lift</div>
<div style="color:#4a5578;margin-top:5px">Load ch71_ab_test.xlsx and report each variant's conversion CI and the CI for the lift (B - A).</div>
</div>

```python
try:    ab = pd.read_excel("../../data/ch71_ab_test.xlsx", sheet_name="Visitors")
except FileNotFoundError: ab = pd.read_excel(BASE+"ch71_ab_test.xlsx", sheet_name="Visitors")
g=ab.groupby("variant").converted.agg(["size","sum"])
pA,nA=g.loc["A","sum"]/g.loc["A","size"],g.loc["A","size"]; pB,nB=g.loc["B","sum"]/g.loc["B","size"],g.loc["B","size"]
seA=np.sqrt(pA*(1-pA)/nA); seB=np.sqrt(pB*(1-pB)/nB)
print(f"A: {pA*100:.2f}% [{(pA-z*seA)*100:.2f}, {(pA+z*seA)*100:.2f}]")
print(f"B: {pB*100:.2f}% [{(pB-z*seB)*100:.2f}, {(pB+z*seB)*100:.2f}]")
diff=pB-pA; se=np.sqrt(pA*(1-pA)/nA+pB*(1-pB)/nB)
print(f"lift {diff*100:+.2f} pts, 95% CI [{(diff-z*se)*100:+.2f}, {(diff+z*se)*100:+.2f}] pts; excludes 0? {diff-z*se>0}")
```
