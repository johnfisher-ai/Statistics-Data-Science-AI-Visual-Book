## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
rng = np.random.default_rng(730)
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
def bootstrap(x, stat, B=10000):
    n=len(x); idx=rng.integers(0,n,size=(B,n)); return stat(np.asarray(x)[idx], axis=1)
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Bootstrap the standard error</div>
<div style="color:#4a5578;margin-top:5px">From a sample of 80 from Normal(100,15), bootstrap the SE of the mean and compare to s/sqrt(n).</div>
</div>

```python
x=rng.normal(100,15,80); bm=bootstrap(x,np.mean)
print(f"bootstrap SE = {bm.std():.3f}, formula s/sqrt(n) = {x.std(ddof=1)/np.sqrt(80):.3f}")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Percentile CI for the mean</div>
<div style="color:#4a5578;margin-top:5px">Build the bootstrap 95% percentile CI for the mean and compare to the t-interval.</div>
</div>

```python
lo,hi=np.percentile(bm,[2.5,97.5]); se=x.std(ddof=1)/np.sqrt(80); t=stats.t.ppf(0.975,79)
print(f"bootstrap CI [{lo:.2f}, {hi:.2f}]")
print(f"t-interval  [{x.mean()-t*se:.2f}, {x.mean()+t*se:.2f}]")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">CI for the median</div>
<div style="color:#4a5578;margin-top:5px">On right-skewed lognormal data (n=150), build a bootstrap 95% CI for the median.</div>
</div>

```python
y=rng.lognormal(3,0.6,150); bmed=bootstrap(y,np.median)
print(f"median {np.median(y):.2f}, bootstrap 95% CI [{np.percentile(bmed,2.5):.2f}, {np.percentile(bmed,97.5):.2f}]")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">CI for a correlation</div>
<div style="color:#4a5578;margin-top:5px">Bootstrap a 95% CI for the correlation between two related variables.</div>
</div>

```python
a=rng.normal(0,1,120); b=0.6*a+rng.normal(0,1,120); data=np.column_stack([a,b])
def corr(arr,axis): return np.array([np.corrcoef(arr[i,:,0],arr[i,:,1])[0,1] for i in range(arr.shape[0])])
n=len(data); idx=rng.integers(0,n,size=(5000,n)); bc=corr(data[idx],1)
print(f"r = {np.corrcoef(a,b)[0,1]:.3f}, bootstrap 95% CI [{np.percentile(bc,2.5):.3f}, {np.percentile(bc,97.5):.3f}]")
```

<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: median salary</div>
<div style="color:#4a5578;margin-top:5px">Load ch73_salaries.xlsx and build a bootstrap 95% CI for the median salary.</div>
</div>

```python
try:    emp = pd.read_excel("../../data/ch73_salaries.xlsx", sheet_name="Employees")
except FileNotFoundError: emp = pd.read_excel(BASE+"ch73_salaries.xlsx", sheet_name="Employees")
sal=emp.annual_salary.values; bmed=bootstrap(sal,np.median,B=20000)
print(f"n={len(sal)}: median ${np.median(sal):,.0f}, bootstrap 95% CI [${np.percentile(bmed,2.5):,.0f}, ${np.percentile(bmed,97.5):,.0f}]")
```
