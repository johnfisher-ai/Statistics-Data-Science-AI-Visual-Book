## ⚙️ Setup

```python
import numpy as np
from scipy import stats
import pandas as pd
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(750)
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Verify the false-positive rate</div>
<div style="color:#4a5578;margin-top:5px">Run 10,000 two-sample t-tests on pure noise (both groups N(0,1), n=30) and confirm ~5% are 'significant'.</div>
</div>

```python
sig=np.mean([stats.ttest_ind(rng.normal(0,1,30),rng.normal(0,1,30)).pvalue<0.05 for _ in range(10000)])
print(f"false-positive rate = {sig:.3f} (target 0.05)")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Estimate power</div>
<div style="color:#4a5578;margin-top:5px">For a true mean of 0.4 (sd 1), n=50, estimate the power of a one-sample t-test at alpha=0.05.</div>
</div>

```python
pw=np.mean([stats.ttest_1samp(rng.normal(0.4,1,50),0).pvalue<0.05 for _ in range(5000)])
print(f"estimated power = {pw:.2f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Power vs sample size</div>
<div style="color:#4a5578;margin-top:5px">Show power rising with n (effect 0.4) for n in [20,50,100,200].</div>
</div>

```python
for n in [20,50,100,200]:
    pw=np.mean([stats.ttest_1samp(rng.normal(0.4,1,n),0).pvalue<0.05 for _ in range(3000)])
    print(f"n={n:>3}: power={pw:.2f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Significance is not importance</div>
<div style="color:#4a5578;margin-top:5px">With n=50,000 and a true effect of 0.01, show the test is 'significant' yet the effect size is negligible.</div>
</div>

```python
x=rng.normal(0.01,1,50000); p=stats.ttest_1samp(x,0).pvalue
print(f"p={p:.3e} (significant) but d={x.mean()/x.std():.3f} (negligible)")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: power of the battery test</div>
<div style="color:#4a5578;margin-top:5px">Load ch75_battery_life.xlsx, run the one-sided t-test of H0: mu=10, and report the power at the observed effect.</div>
</div>

```python
try:    bat = pd.read_excel("../../data/ch75_battery_life.xlsx", sheet_name="Batteries")
except FileNotFoundError: bat = pd.read_excel(BASE+"ch75_battery_life.xlsx", sheet_name="Batteries")
from scipy.stats import nct, t as tdist
x=bat.life_hours; n=len(x); d=(x.mean()-10)/x.std(ddof=1)
res=stats.ttest_1samp(x,10)
power=nct.cdf(tdist.ppf(0.05,n-1), n-1, d*np.sqrt(n))
print(f"one-sided p={res.pvalue/2:.4f}, d={d:.2f}, power={power:.2f}")
```
