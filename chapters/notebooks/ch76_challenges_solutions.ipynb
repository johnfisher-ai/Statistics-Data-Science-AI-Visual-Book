## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(760)
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">One-proportion z-test</div>
<div style="color:#4a5578;margin-top:5px">In 1,000 trials you see 470 successes. Test H0: p = 0.5 (two-sided).</div>
</div>

```python
s,n,p0=470,1000,0.5; ph=s/n; se=np.sqrt(p0*(1-p0)/n); z=(ph-p0)/se
print(f"z={z:.2f}, p={2*(1-stats.norm.cdf(abs(z))):.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Two-proportion z-test</div>
<div style="color:#4a5578;margin-top:5px">Group A: 90/500. Group B: 130/520. Test H0: pA = pB.</div>
</div>

```python
sA,nA,sB,nB=90,500,130,520; pA,pB=sA/nA,sB/nB; pool=(sA+sB)/(nA+nB)
se=np.sqrt(pool*(1-pool)*(1/nA+1/nB)); z=(pB-pA)/se
print(f"pA={pA:.3f}, pB={pB:.3f}, z={z:.2f}, p={2*(1-stats.norm.cdf(abs(z))):.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Large-sample z-test for a mean</div>
<div style="color:#4a5578;margin-top:5px">A sample of 500 has mean 51 and sd 10. Test H0: mu = 50 (two-sided).</div>
</div>

```python
xbar,s,n,mu0=51,10,500,50; se=s/np.sqrt(n); z=(xbar-mu0)/se
print(f"z={z:.2f}, p={2*(1-stats.norm.cdf(abs(z))):.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Check the validity condition</div>
<div style="color:#4a5578;margin-top:5px">For n=40 and p0=0.05, is the normal (z) approximation valid? Explain with the success-failure rule.</div>
</div>

```python
n,p0=40,0.05; print(f"n*p0={n*p0}, n*(1-p0)={n*(1-p0)}")
print("n*p0 = 2 < 10 -> normal approximation NOT valid; use an exact binomial test instead")
```
