## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
from scipy.stats import studentized_range
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(780)
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">One-way ANOVA</div>
<div style="color:#4a5578;margin-top:5px">Three groups N(20,5,n=30), N(22,5,n=30), N(26,5,n=30). Run a one-way ANOVA.</div>
</div>

```python
g=[rng.normal(m,5,30) for m in (20,22,26)]; F,p=stats.f_oneway(*g)
print(f"F={F:.2f}, p={p:.2e}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Effect size eta-squared</div>
<div style="color:#4a5578;margin-top:5px">For those groups compute eta^2 = SS_between / SS_total.</div>
</div>

```python
allv=np.concatenate(g); grand=allv.mean()
ssb=sum(len(x)*(x.mean()-grand)**2 for x in g); sst=((allv-grand)**2).sum()
print(f"eta^2 = {ssb/sst:.3f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">ANOVA equals the t-test for 2 groups</div>
<div style="color:#4a5578;margin-top:5px">Show that for exactly 2 groups, F = t^2.</div>
</div>

```python
a=rng.normal(10,3,40); b=rng.normal(12,3,40)
t=stats.ttest_ind(a,b,equal_var=True).statistic; F=stats.f_oneway(a,b).statistic
print(f"t^2={t**2:.3f}, F={F:.3f} -> equal")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Tukey HSD by hand</div>
<div style="color:#4a5578;margin-top:5px">For the three groups in #1, find which pairs differ (alpha=0.05).</div>
</div>

```python
k=len(g); N=sum(len(x) for x in g); dfw=N-k
msw=sum(((x-x.mean())**2).sum() for x in g)/dfw; q=studentized_range.ppf(0.95,k,dfw)
for i in range(k):
    for j in range(i+1,k):
        diff=g[i].mean()-g[j].mean(); hsd=q*np.sqrt(msw/2*(1/len(g[i])+1/len(g[j])))
        print(f"  {i} vs {j}: diff={diff:+.2f}, sig={abs(diff)>hsd}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: carriers</div>
<div style="color:#4a5578;margin-top:5px">Load ch78_carrier_delivery.xlsx and run the one-way ANOVA across carriers.</div>
</div>

```python
try:    d = pd.read_excel("../../data/ch78_carrier_delivery.xlsx", sheet_name="Shipments")
except FileNotFoundError: d = pd.read_excel(BASE+"ch78_carrier_delivery.xlsx", sheet_name="Shipments")
groups=[x.delivery_hours.values for _,x in d.groupby("carrier")]
F,p=stats.f_oneway(*groups); print(f"F={F:.2f}, p={p:.2e}")
```
