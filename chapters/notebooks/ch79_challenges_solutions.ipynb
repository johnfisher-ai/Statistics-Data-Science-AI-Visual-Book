## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(790)
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Goodness-of-fit for a die</div>
<div style="color:#4a5578;margin-top:5px">Test whether [18,22,16,20,14,10] (100 rolls) is consistent with a fair die.</div>
</div>

```python
obs=np.array([18,22,16,20,14,10]); exp=np.full(6,obs.sum()/6)
r=stats.chisquare(obs,exp); print(f"chi2={r.statistic:.2f}, df=5, p={r.pvalue:.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Goodness-of-fit vs a claim</div>
<div style="color:#4a5578;margin-top:5px">Observed [330,290,180,100]; claim 40/30/20/10. Test it.</div>
</div>

```python
obs=np.array([330,290,180,100]); exp=np.array([.4,.3,.2,.1])*obs.sum()
r=stats.chisquare(obs,exp); print(f"chi2={r.statistic:.2f}, p={r.pvalue:.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Test of independence</div>
<div style="color:#4a5578;margin-top:5px">For the table [[90,60,50],[40,80,70]], test independence and report dof.</div>
</div>

```python
chi2,p,dof,exp=stats.chi2_contingency([[90,60,50],[40,80,70]])
print(f"chi2={chi2:.2f}, dof={dof}, p={p:.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Cramer's V</div>
<div style="color:#4a5578;margin-top:5px">For the table above, compute Cramer's V as an effect size.</div>
</div>

```python
t=np.array([[90,60,50],[40,80,70]]); chi2,_,_,_=stats.chi2_contingency(t)
V=np.sqrt(chi2/(t.sum()*(min(t.shape)-1))); print(f"Cramer\u2019s V = {V:.3f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: channel by region</div>
<div style="color:#4a5578;margin-top:5px">Load ch79_customer_channel.xlsx and test independence of region and preferred_channel.</div>
</div>

```python
try:    d = pd.read_excel("../../data/ch79_customer_channel.xlsx", sheet_name="Customers")
except FileNotFoundError: d = pd.read_excel(BASE+"ch79_customer_channel.xlsx", sheet_name="Customers")
ct=pd.crosstab(d.region,d.preferred_channel); chi2,p,dof,_=stats.chi2_contingency(ct)
print(f"chi2={chi2:.2f}, dof={dof}, p={p:.2e}")
```
