## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(800)
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Mann-Whitney U</div>
<div style="color:#4a5578;margin-top:5px">Compare lognormal(1.0,0.6,n=50) and lognormal(0.6,0.6,n=50) with the rank-sum test.</div>
</div>

```python
a=rng.lognormal(1.0,0.6,50); b=rng.lognormal(0.6,0.6,50)
u,p=stats.mannwhitneyu(a,b,alternative="two-sided"); print(f"U={u:.1f}, p={p:.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Outlier robustness</div>
<div style="color:#4a5578;margin-top:5px">Add one outlier of 500 to a clean N(50,8,40) sample; compare how the t-test p and Mann-Whitney p vs N(52,8,40) change.</div>
</div>

```python
clean=rng.normal(50,8,40); other=rng.normal(52,8,40); dirty=clean.copy(); dirty[0]=500
print(f"t-test (dirty) p={stats.ttest_ind(dirty,other).pvalue:.3f}")
print(f"Mann-Whitney (dirty) p={stats.mannwhitneyu(dirty,other).pvalue:.3f}  (ranks ignore the outlier\u2019s size)")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Kruskal-Wallis</div>
<div style="color:#4a5578;margin-top:5px">Compare three skewed groups lognormal at mu=1.0,1.2,1.5 (n=40 each).</div>
</div>

```python
g=[rng.lognormal(m,0.5,40) for m in (1.0,1.2,1.5)]; H,p=stats.kruskal(*g)
print(f"H={H:.2f}, p={p:.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Wilcoxon signed-rank</div>
<div style="color:#4a5578;margin-top:5px">Paired: before=lognormal(1.2,0.4,30), after=before*uniform(0.7,0.95). Test the paired drop.</div>
</div>

```python
before=rng.lognormal(1.2,0.4,30); after=before*rng.uniform(0.7,0.95,30)
w,p=stats.wilcoxon(after,before); print(f"W={w:.1f}, p={p:.4f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: ticket times</div>
<div style="color:#4a5578;margin-top:5px">Load ch80_ticket_times.xlsx; run Mann-Whitney (Alpha vs Bravo) and Kruskal-Wallis (by priority).</div>
</div>

```python
try:    d = pd.read_excel("../../data/ch80_ticket_times.xlsx", sheet_name="Tickets")
except FileNotFoundError: d = pd.read_excel(BASE+"ch80_ticket_times.xlsx", sheet_name="Tickets")
A=d[d.team=="Alpha"].resolution_hours; B=d[d.team=="Bravo"].resolution_hours
print(f"Mann-Whitney p={stats.mannwhitneyu(A,B).pvalue:.4f}")
g=[x.resolution_hours.values for _,x in d.groupby("priority")]
print(f"Kruskal-Wallis p={stats.kruskal(*g).pvalue:.2e}")
```
