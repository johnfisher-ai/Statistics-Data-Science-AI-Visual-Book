## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(810)
try:    d = pd.read_excel("../../data/choosing-the-right-test--clinic_visits.xlsx", sheet_name="Visits")
except FileNotFoundError: d = pd.read_excel(BASE+"choosing-the-right-test--clinic_visits.xlsx", sheet_name="Visits")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Name the test</div>
<div style="color:#4a5578;margin-top:5px">Outcome = numeric, 2 paired groups, roughly normal. Which test? Then run it on BP before/after.</div>
</div>

```python
print("paired t-test")
print(f"p={stats.ttest_rel(d.sysbp_after,d.sysbp_before).pvalue:.2e}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Name the test</div>
<div style="color:#4a5578;margin-top:5px">Outcome = numeric, 2 independent groups. Which test? Run it on BP drop by treatment.</div>
</div>

```python
drop=d.sysbp_before-d.sysbp_after
print("two-sample (Welch) t-test")
print(f"p={stats.ttest_ind(drop[d.treatment=="new"],drop[d.treatment=="standard"],equal_var=False).pvalue:.2e}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Name the test</div>
<div style="color:#4a5578;margin-top:5px">Two categorical variables. Which test? Run it on treatment x satisfied.</div>
</div>

```python
print("chi-square test of independence")
print(f"p={stats.chi2_contingency(pd.crosstab(d.treatment,d.satisfied))[1]:.3f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Name the test</div>
<div style="color:#4a5578;margin-top:5px">Numeric outcome, 3 groups, but heavily skewed. Which test? Run it on wait_minutes by clinic.</div>
</div>

```python
print("Kruskal-Wallis (skewed -> nonparametric ANOVA)")
g=[x.wait_minutes.values for _,x in d.groupby("clinic")]
print(f"p={stats.kruskal(*g).pvalue:.3f}")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#4338ca;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">One proportion vs a target</div>
<div style="color:#4a5578;margin-top:5px">Is the overall satisfaction rate different from 0.70? Pick and run the test.</div>
</div>

```python
s=d.satisfied; n=len(s); ph=s.mean(); p0=0.70; z=(ph-p0)/np.sqrt(p0*(1-p0)/n)
print(f"one-proportion z-test: phat={ph:.3f}, z={z:.2f}, p={2*(1-stats.norm.cdf(abs(z))):.3f}")
```
