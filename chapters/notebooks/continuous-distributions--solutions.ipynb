## ⚙️ Setup

```python
import numpy as np
from scipy import stats
rng = np.random.default_rng(381)
print("ready")
```

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 1 · UNIFORM</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Flat probability</div>
<div style="color:#4a5578;margin-top:6px">A bus is equally likely to arrive at any minute in a 10-minute window, X ~ Uniform(0, 10). Find P(X < 3) and the mean wait.</div>
</div>

```python
print(f"P(X < 3)  = 3/10 = {stats.uniform.cdf(3, 0, 10):.3f}")
print(f"mean      = (a+b)/2 = {stats.uniform.mean(0, 10):.1f}")
```

**Answer:** for a uniform, probability is proportional to length, so P(X < 3) = 3/10 = **0.30**, and the mean is the midpoint (0 + 10)/2 = **5** minutes.

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0891b2;letter-spacing:1px">CHALLENGE 2 · EXPONENTIAL</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Waiting times</div>
<div style="color:#4a5578;margin-top:6px">Buses arrive on average every 10 minutes (exponential). Find P(wait > 15 min) and P(wait < 5 min).</div>
</div>

```python
scale = 10        # mean = 1/lambda = 10
print(f"P(wait > 15) = e^(-15/10) = {1 - stats.expon.cdf(15, scale=scale):.4f}")
print(f"P(wait < 5)  = 1 - e^(-5/10) = {stats.expon.cdf(5, scale=scale):.4f}")
```

**Answer:** P(wait > 15) = e&#8315;&#185;&#183;&#8309; &approx; **0.223**, and P(wait < 5) = 1 &minus; e&#8315;&#8304;&#183;&#8309; &approx; **0.393**. The exponential has a long right tail, so long waits, while less likely, are far from rare.

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0d9488;letter-spacing:1px">CHALLENGE 3 · NORMAL</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Probabilities under the bell</div>
<div style="color:#4a5578;margin-top:6px">IQ scores are modeled as X ~ Normal(100, 15). Find P(X < 115) and P(85 < X < 115).</div>
</div>

```python
mu, sd = 100, 15
print(f"P(X < 115)        = {stats.norm.cdf(115, mu, sd):.4f}")
p_mid = stats.norm.cdf(115, mu, sd) - stats.norm.cdf(85, mu, sd)
print(f"P(85 < X < 115)   = {p_mid:.4f}")
```

**Answer:** 115 is one standard deviation above the mean, so P(X < 115) = &Phi;(1) &approx; **0.8413**. The range 85 to 115 is exactly &#956; &plusmn; 1&#963;, so P(85 < X < 115) &approx; **0.6827**, the famous 68% from the empirical rule (Chapter 39).

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#db2777;letter-spacing:1px">CHALLENGE 4 · QUANTILES</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Reading the CDF backward</div>
<div style="color:#4a5578;margin-top:6px">For the same IQ distribution N(100, 15), find the 90th percentile, the score that 90% of people fall below.</div>
</div>

```python
mu, sd = 100, 15
p90 = stats.norm.ppf(0.90, mu, sd)
print(f"90th percentile = {p90:.2f}")
```

**Answer:** the 90th percentile is &#956; + z&#8320;.&#8313;&#8320;&#183;&#963; = 100 + 1.2816(15) &approx; **119.2**. The percent-point function (inverse CDF) turns a probability back into a value, the reverse of asking "what fraction is below x".

<div style="background:#ecfdf5;border-left:5px solid #0d9488;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0f766e;letter-spacing:1px">CHALLENGE 5 · FIT A GAUSSIAN</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Maximum likelihood</div>
<div style="color:#4a5578;margin-top:6px">Given a sample of measurements, estimate the parameters of a normal distribution by maximum likelihood.</div>
</div>

```python
sample = rng.normal(50, 12, size=300)
mu_hat, sigma_hat = sample.mean(), sample.std()
print(f"MLE mu_hat    = {mu_hat:.2f}")
print(f"MLE sigma_hat = {sigma_hat:.2f}")
print("Assuming Gaussian errors, minimizing squared error == maximizing this likelihood.")
```

**Answer:** the maximum-likelihood estimates are simply the sample mean and sample standard deviation (here near 50 and 12, the values we generated from). The deeper point: assuming normally distributed errors makes **least-squares fitting the maximum-likelihood choice**, which is why squared-error loss is everywhere in regression and deep learning.
