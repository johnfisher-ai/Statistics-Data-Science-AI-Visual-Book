## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
import statsmodels.api as sm
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(900)
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Pearson r and r-squared</div>
<div style="color:#4a5578;margin-top:5px">For a pair with rho=0.6, compute Pearson r and report r^2 as a percentage.</div>
</div>

```python
x=rng.normal(0,1,400); y=0.6*x+np.sqrt(1-0.36)*rng.normal(0,1,400)
r,_=stats.pearsonr(x,y); print(f"r={r:.2f}, r^2={r**2:.0%}")
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Spearman beats Pearson on a curve</div>
<div style="color:#4a5578;margin-top:5px">For y=x^3, show Spearman rho is near 1 but Pearson r is lower.</div>
</div>

```python
x=np.sort(rng.uniform(-3,3,200)); y=x**3+rng.normal(0,1,200)
print(f"Pearson={stats.pearsonr(x,y)[0]:.2f}, Spearman={stats.spearmanr(x,y)[0]:.2f}")
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Outlier robustness</div>
<div style="color:#4a5578;margin-top:5px">Add one outlier to a clean linear pair; compare Pearson and Spearman.</div>
</div>

```python
a=rng.normal(0,1,100); b=a+rng.normal(0,0.3,100); a[0],b[0]=8,-8
print(f"Pearson={stats.pearsonr(a,b)[0]:+.2f}, Spearman={stats.spearmanr(a,b)[0]:+.2f}")
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Partial correlation</div>
<div style="color:#4a5578;margin-top:5px">x and y both driven by z. Show the partial correlation controlling for z collapses.</div>
</div>

```python
z=rng.normal(0,1,300); x=z+rng.normal(0,0.5,300); y=z+rng.normal(0,0.5,300)
def resid(v,w): return sm.OLS(v,sm.add_constant(w)).fit().resid
print(f"raw={stats.pearsonr(x,y)[0]:.2f}, partial(|z)={stats.pearsonr(resid(x,z),resid(y,z))[0]:.2f}")
```
