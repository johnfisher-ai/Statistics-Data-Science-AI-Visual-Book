## ⚙️ Setup

```python
import itertools
from math import factorial, perm, comb
print("ready")
```

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#4f46e5;letter-spacing:1px">CHALLENGE 1 · MULTIPLICATION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Count the menus</div>
<div style="color:#4a5578;margin-top:6px">A cafe offers 4 mains, 3 sides, and 5 drinks. How many distinct meals (one of each)? What if you also pick one of 2 desserts?</div>
</div>

```python
meals = 4 * 3 * 5
with_dessert = 4 * 3 * 5 * 2
print(f"meals = 4 x 3 x 5 = {meals}")
print(f"with a dessert = 4 x 3 x 5 x 2 = {with_dessert}")
```

**Answer:** each independent choice multiplies, so 4 x 3 x 5 = 60 meals, and adding a 2-way dessert choice doubles it to 120. The multiplication principle scales effortlessly: every new decision just multiplies in its own number of options.

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#2563eb;letter-spacing:1px">CHALLENGE 2 · PERMUTATIONS</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Order matters</div>
<div style="color:#4a5578;margin-top:6px">From a club of 10 members, in how many ways can you choose a President, a Vice-President, and a Treasurer (distinct roles)?</div>
</div>

```python
n, r = 10, 3
print(f"P({n},{r}) = {n}!/({n}-{r})! = {perm(n,r)}")
print("check by enumeration:", len(list(itertools.permutations(range(n), r))))
```

**Answer:** the three roles are different, so order matters and this is a permutation: P(10,3) = 10 x 9 x 8 = 720. (The president has 10 candidates, then 9 remain for VP, then 8 for treasurer.) If the three positions were interchangeable, it would instead be a combination.

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#059669;letter-spacing:1px">CHALLENGE 3 · COMBINATIONS</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Order ignored</div>
<div style="color:#4a5578;margin-top:6px">From the same 10 members, how many ways to choose a 3-person committee where all roles are equal? How many handshakes occur if all 10 shake hands once?</div>
</div>

```python
print(f"committees = C(10,3) = {comb(10,3)}")
print(f"handshakes = C(10,2) = {comb(10,2)}   (each handshake is a pair)")
```

**Answer:** a committee has no internal order, so it is a combination: C(10,3) = 120, exactly the permutation 720 divided by 3! = 6. Handshakes are unordered pairs, so C(10,2) = 45. Whenever you are selecting a group rather than arranging one, reach for the combination.

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#db2777;letter-spacing:1px">CHALLENGE 4 · WHICH ONE?</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Permutation or combination?</div>
<div style="color:#4a5578;margin-top:6px">For each, decide whether order matters and compute it: (a) a 4-character password from 26 letters with no repeats; (b) choosing 4 pizza toppings from 12.</div>
</div>

```python
pw = perm(26,4)        # order matters: ABCD differs from DCBA
pizza = comb(12,4)     # order does not matter on a pizza
print(f"(a) password, order MATTERS  -> P(26,4) = {pw:,}")
print(f"(b) toppings, order IGNORED  -> C(12,4) = {pizza}")
```

**Answer:** (a) a password is ordered, ABCD is not DCBA, so it is a permutation: P(26,4) = 358,800. (b) pizza toppings are an unordered set, so it is a combination: C(12,4) = 495. The deciding question is always the same: would reordering the same items count as a different outcome?

<div style="background:#eef2ff;border-left:5px solid #4f46e5;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#d97706;letter-spacing:1px">CHALLENGE 5 · COUNTING FOR PROBABILITY</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Favorable over total</div>
<div style="color:#4a5578;margin-top:6px">From a standard 52-card deck, what is the probability that a 5-card hand is exactly four-of-a-kind (four cards of one rank plus any fifth card)?</div>
</div>

```python
hands = comb(52,5)
# 13 ranks for the quad, then the 5th card is any of the remaining 48
four_kind = 13 * comb(48,1)
p = four_kind / hands
print(f"total hands      = C(52,5) = {hands:,}")
print(f"four-of-a-kind   = 13 x 48 = {four_kind:,}")
print(f"P(four-of-a-kind)= {four_kind}/{hands} = {p:.8f}  (about 1 in {round(1/p):,})")
```

**Answer:** count the favorable hands, choose the rank for the four-of-a-kind (13 ways), then any of the 48 remaining cards as the fifth, giving 13 x 48 = 624 hands out of C(52,5) = 2,598,960. So P ≈ 0.00024, about 1 in 4,165. Counting the favorable and total outcomes is the whole job; the probability is their ratio.
