## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
rng=np.random.default_rng(890)
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Covariance from the formula</div>
<div style="color:#4a5578;margin-top:5px">For x=[1,2,3,4,5], y=[2,4,5,4,5], compute the covariance as the mean of deviation products and check with numpy.</div>
</div>

```python
x=np.array([1,2,3,4,5]); y=np.array([2,4,5,4,5])
cov=((x-x.mean())*(y-y.mean())).mean()
print(f"formula (ddof=0) = {cov:.3f}, numpy = {np.cov(x,y,ddof=0)[0,1]:.3f}")
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Sign of covariance</div>
<div style="color:#4a5578;margin-top:5px">Generate a negatively related pair and confirm the covariance is negative.</div>
</div>

```python
a=rng.normal(0,1,200); b=-0.7*a+rng.normal(0,0.7,200)
print(f"cov = {np.cov(a,b)[0,1]:.3f} (negative as expected)")
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Units change covariance, not correlation</div>
<div style="color:#4a5578;margin-top:5px">Show cov(x, 100*y) = 100*cov(x, y) but the correlation is unchanged.</div>
</div>

```python
x=rng.normal(0,1,300); y=0.6*x+rng.normal(0,0.8,300)
print(f"cov(x,y)={np.cov(x,y)[0,1]:.3f}, cov(x,100y)={np.cov(x,100*y)[0,1]:.3f}")
print(f"corr(x,y)={np.corrcoef(x,y)[0,1]:.3f}, corr(x,100y)={np.corrcoef(x,100*y)[0,1]:.3f}")
```

<div style="background:#faf5ff;border-left:5px solid #9333ea;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#7e22ce;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Standardize to correlation</div>
<div style="color:#4a5578;margin-top:5px">Compute the correlation as cov/(sx*sy) and confirm it matches np.corrcoef.</div>
</div>

```python
print(f"by hand = {np.cov(x,y)[0,1]/(x.std(ddof=1)*y.std(ddof=1)):.4f}, corrcoef = {np.corrcoef(x,y)[0,1]:.4f}")
```
