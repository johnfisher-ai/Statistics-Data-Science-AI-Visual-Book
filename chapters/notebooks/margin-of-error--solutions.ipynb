## ⚙️ Setup

```python
import numpy as np, pandas as pd
from scipy import stats
z = stats.norm.ppf(0.975)
BASE="https://raw.githubusercontent.com/johnfisher-ai/Statistics-Data-Science-AI-Visual-Book/main/data/"
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 1</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Compute a margin of error</div>
<div style="color:#4a5578;margin-top:5px">A poll of 1,200 finds 47% support. Compute the 95% margin of error.</div>
</div>

```python
p,n=0.47,1200; moe=z*np.sqrt(p*(1-p)/n)
print(f"MoE = +/- {moe*100:.2f} pts -> {p*100:.0f}% +/- {moe*100:.1f}%")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 2</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Worst-case margin</div>
<div style="color:#4a5578;margin-top:5px">Not knowing p, compute the WORST-CASE 95% margin for n=1,200 (use p=0.5).</div>
</div>

```python
print(f"worst case p=0.5: MoE = +/- {z*np.sqrt(0.25/1200)*100:.2f} pts")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 3</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Confidence widens the margin</div>
<div style="color:#4a5578;margin-top:5px">For n=1,000 and p=0.5, tabulate the margin at 90%, 95%, 99% confidence.</div>
</div>

```python
for c in [0.90,0.95,0.99]:
    zz=stats.norm.ppf(0.5+c/2); print(f"{c:.0%}: +/- {zz*np.sqrt(0.25/1000)*100:.2f} pts")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 4</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Quadruple for half</div>
<div style="color:#4a5578;margin-top:5px">Confirm that going from n=1,000 to n=4,000 halves the margin of error.</div>
</div>

```python
m1=z*np.sqrt(0.25/1000); m4=z*np.sqrt(0.25/4000)
print(f"n=1000: +/-{m1*100:.2f} pts | n=4000: +/-{m4*100:.2f} pts | ratio {m4/m1:.2f}")
```

<div style="background:#eef2ff;border-left:5px solid #2563eb;border-radius:10px;padding:14px 18px;font-family:Inter,sans-serif">
<span style="font-size:12px;font-weight:800;color:#1d4ed8;letter-spacing:1px">CHALLENGE 5</span>
<div style="font-size:18px;font-weight:800;color:#1a2138;margin-top:3px">Real data: survey margin</div>
<div style="color:#4a5578;margin-top:5px">Load margin-of-error--customer_survey.xlsx and report the recommend rate with its 95% margin of error.</div>
</div>

```python
try:    s = pd.read_excel("../../data/margin-of-error--customer_survey.xlsx", sheet_name="Responses")
except FileNotFoundError: s = pd.read_excel(BASE+"margin-of-error--customer_survey.xlsx", sheet_name="Responses")
p=s.would_recommend.mean(); n=len(s); moe=z*np.sqrt(p*(1-p)/n)
print(f"n={n}: {p*100:.0f}% would recommend, +/- {moe*100:.1f} pts (95%)")
```
