## ⚙️ Setup

```python
import numpy as np
import pandas as pd
print("✅ Ready.")
```

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#0891b2;letter-spacing:1px">CHALLENGE 1 · RANGE & IQR</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Spread of a small dataset</div>
<div style="color:#4a5578;margin-top:6px">For [12, 15, 14, 10, 18, 20, 13, 16], find the range, the quartiles Q1 and Q3, and the IQR. (This is the five-number summary minus the median.)</div>
</div>

```python
x = pd.Series([12, 15, 14, 10, 18, 20, 13, 16])
q1, med, q3 = x.quantile([.25, .5, .75])
print(f"Min {x.min()}, Q1 {q1}, Median {med}, Q3 {q3}, Max {x.max()}")
print(f"Range = {x.max()-x.min()}")
print(f"IQR   = Q3 - Q1 = {q3-q1}")
```

**Answer:** Range = **10** (20−10). With NumPy/pandas quartiles, Q1 = **13.25**, Q3 = **16.5**, so IQR = **3.25**. (Different software may report slightly different quartiles, that is normal.)

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#7c3aed;letter-spacing:1px">CHALLENGE 2 · VARIANCE & SD</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Build them by hand</div>
<div style="color:#4a5578;margin-top:6px">For [5, 7, 3, 9, 6], compute the deviations from the mean, then the SAMPLE variance (÷ n−1) and sample standard deviation. Confirm with NumPy.</div>
</div>

```python
x = np.array([5, 7, 3, 9, 6])
mean = x.mean()
dev  = x - mean
var_samp = np.sum(dev**2)/(len(x)-1)
sd_samp  = np.sqrt(var_samp)
print(f"mean = {mean}")
print(f"deviations = {dev} (sum {dev.sum()})")
print(f"sample variance = {var_samp:.2f},  sample SD = {sd_samp:.2f}")
print(f"NumPy: np.std(x, ddof=1) = {np.std(x, ddof=1):.2f}")
```

**Answer:** mean = 6, deviations [−1, 1, −3, 3, 0] (sum 0, as always). Sample variance = **5.0**, sample SD ≈ **2.24**. Dividing by n−1 = 4 (not 5) gives the unbiased sample estimate.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#d97706;letter-spacing:1px">CHALLENGE 3 · THE 68-95-99.7 RULE</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Use the standard deviation as a ruler</div>
<div style="color:#4a5578;margin-top:6px">Test scores are roughly bell-shaped with mean 500 and SD 100. (a) About what range holds the middle 95%? (b) Roughly what % of scores are above 700?</div>
</div>

```python
mean, sd = 500, 100
low, high = mean - 2*sd, mean + 2*sd
print(f"(a) ~95% of scores fall in {low} to {high}")
# 700 is 2 SD above the mean. ~95% are within +/-2 SD, so ~5% are outside,
# split evenly -> ~2.5% above +2 SD.
print("(b) 700 is +2 SD; about 2.5% of scores are above 700.")
```

**Answer:** (a) about **300 to 700** (mean ± 2 SD). (b) 700 is exactly 2 SD above the mean; ~5% lie beyond ±2 SD, so ~**2.5%** score above 700. (Holds because the data is approximately normal.)

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#2563eb;letter-spacing:1px">CHALLENGE 4 · COEFFICIENT OF VARIATION</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Compare spread across scales</div>
<div style="color:#4a5578;margin-top:6px">Stock A has mean monthly return 8% with SD 4%. Stock B has mean 2% with SD 1.5%. Which is RELATIVELY more variable? Use the coefficient of variation.</div>
</div>

```python
for name, mean, sd in [("Stock A", 8, 4), ("Stock B", 2, 1.5)]:
    cv = sd/mean*100
    print(f"{name}: CV = {sd}/{mean} = {cv:.1f}%")
```

**Answer:** Stock A CV = 50%, Stock B CV = **75%**. Even though B's SD (1.5) is smaller in absolute terms, relative to its tiny mean it is the more variable, the higher CV. CV lets you compare spread fairly across different scales.

<div style="background:#fef3e2;border-left:5px solid #d97706;border-radius:10px;padding:16px 20px;font-family:Inter,sans-serif">
<span style="font-size:13px;font-weight:800;color:#059669;letter-spacing:1px">CHALLENGE 5 · ROBUSTNESS</span>
<div style="font-size:21px;font-weight:800;color:#1a2138;margin-top:4px">Which measure survives an outlier?</div>
<div style="color:#4a5578;margin-top:6px">Start with [20, 22, 21, 23, 19]. Add the value 200. Recompute the standard deviation and the IQR, and say which measure of spread is the robust one.</div>
</div>

```python
base = pd.Series([20, 22, 21, 23, 19])
out  = pd.concat([base, pd.Series([200])], ignore_index=True)
def iqr(s): return s.quantile(.75) - s.quantile(.25)
print(f"Before: SD {base.std(ddof=1):.1f}, IQR {iqr(base):.1f}")
print(f"After : SD {out.std(ddof=1):.1f}, IQR {iqr(out):.1f}")
```

**Answer:** the **SD** leaps from about 1.6 to roughly 73, while the **IQR** stays small. The IQR is the robust measure: like the median, it ignores how extreme an outlier is. Report median + IQR for skewed or outlier-prone data.
